# Wi-Fi Distance Study (Results & Plots)

This notebook visualizes output from the ns-3 Wi-Fi distance simulation.

**Youâ€™ll find:**
- Data preview & sanity checks  
- **Throughput vs Distance** plot  
- **Average Delay vs Distance** plot  
- Packet-loss snapshot + simple correlations  


```python
# Configuration
CSV_PATH = 'results.csv'   # notebook lives next to the CSV

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Helpful error if CSV is missing
if not os.path.exists(CSV_PATH):
    raise FileNotFoundError(
        f"results.csv not found at '{CSV_PATH}'. "
        "Move the file here (e.g., analysis/results.csv) and try again."
    )

df = pd.read_csv(CSV_PATH)

# Normalize column names
df.columns = [c.strip().lower() for c in df.columns]

required = {'distance_m','throughput_mbps','avg_delay_ms','packet_loss_percent'}
missing = required - set(df.columns)
if missing:
    raise ValueError(
        f"CSV is missing required columns: {sorted(missing)}\n"
        f"Found columns: {list(df.columns)}"
    )

# Sort & de-duplicate by distance
df = df.sort_values('distance_m').drop_duplicates('distance_m', keep='first').reset_index(drop=True)

print(f"Loaded {len(df)} rows from '{CSV_PATH}'.")
df.head()
```

## Summary Stats
Quick overview to confirm values look reasonable.


```python
summary = df.describe().T[['mean','std','min','max']]
summary.rename(columns={'mean':'Mean','std':'Std','min':'Min','max':'Max'}, inplace=True)
summary
```

## Plot: Throughput vs Distance


```python
plt.figure()
plt.plot(df['distance_m'], df['throughput_mbps'], marker='o')
plt.xlabel('Distance (m)')
plt.ylabel('Throughput (Mbps)')
plt.title('Wi-Fi Throughput vs Distance')
plt.grid(True)
plt.savefig('throughput_vs_distance.png', dpi=180, bbox_inches='tight')
plt.show()
```

## Plot: Average Delay vs Distance


```python
plt.figure()
plt.plot(df['distance_m'], df['avg_delay_ms'], marker='o')
plt.xlabel('Distance (m)')
plt.ylabel('Average One-Way Delay (ms)')
plt.title('Wi-Fi Average Delay vs Distance')
plt.grid(True)
plt.savefig('delay_vs_distance.png', dpi=180, bbox_inches='tight')
plt.show()
```

## Notes & Interpretation

### 1. Throughput vs Distance
- At **short distances (5â€“20 m)**, throughput is fairly stable around **12â€“12.5 Mbps**.  
- Beyond **20 m**, throughput **drops sharply**: ~10.7 Mbps at 35 m and ~9 Mbps at 50 m.  
- This is expected: as distance increases, the **signal-to-noise ratio (SNR)** falls â†’ more retransmissions and lower PHY data rates â†’ reduced throughput.

---

### 2. Delay vs Distance
- **Short distances (5â€“20 m):** Delay is ~80 ms, relatively low and stable.  
- **Medium distance (30â€“35 m):** Delay rises toward ~90 ms.  
- **Far distance (50 m):** Delay spikes above **100 ms**.  
- At longer ranges, **packets take longer to be delivered and acknowledged**, mostly due to **MAC-layer retransmissions**.

---

### 3. Combined Story
- **Stable region:** Up to ~20 m, performance is nearly flat â†’ Wi-Fi provides consistent throughput and low delay.  
- **Degradation region:** Beyond 20 m, both throughput and delay degrade quickly â†’ the **link quality margin is shrinking**.  
- **Implication:** In real deployments, Wi-Fi coverage planning must account for these thresholds, as user experience can drop disproportionately after a certain distance.

---

## Conclusion
- The simulation demonstrates a core wireless principle:  
  - **Throughput decreases** with distance.  
  - **Delay increases** with distance.
